```python
def med(a, b):
    m = len(a)
    n = len(b)

    dp = [[0]*(n+1) for i in range(m+1)]

    for i in range(m+1):
        dp[i][0] = i

    for j in range(n+1):
        dp[0][j] = j

    for i in range(1, m+1):
        for j in range(1, n+1):
            if a[i-1] == b[j-1]:
                dp[i][j] = dp[i-1][j-1]
            else:
                dp[i][j] = 1 + min(
                    dp[i-1][j],      # delete
                    dp[i][j-1],      # insert
                    dp[i-1][j-1]     # replace
                )

    return dp[m][n]


tests = [
    ("kitten","sitting"),
    ("flaw","lawn"),
    ("intention","execution"),
    ("apple","aple"),
    ("recieve","receive")
]

for x, y in tests:
    print("MED between", x, "and", y, ":", med(x, y))
```
