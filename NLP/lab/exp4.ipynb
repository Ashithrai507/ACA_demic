```python
class TDP:
    def __init__(s,x): 
        s.x=x; s.i=0
    def m(s,c):
        if s.i<len(s.x) and s.x[s.i]==c: s.i+=1; return 1
    def A(s): s.m('b'); return 1
    def B(s): 
        return s.m('c')
    def parse(s): 
        return s.m('a') and s.A() and s.B() and s.i==len(s.x)

print("Top-Down parsing Result:",TDP("abc").parse())
```

```python
class BUP:
    def __init__(s,x): s.x=list(x); s.st=[]
    def parse(s):
        while s.x or len(s.st)>1:
            if s.x: s.st.append(s.x.pop(0))   # shift
            if ''.join(s.st[-2:])=='bc': s.st[-2:]=['A']   # reduce A→b
            if ''.join(s.st[-2:])=='aA': s.st[-2:]=['S']   # reduce S→aAB (partial)
        return s.st==['S']

print("Bottom-Up Parsing Result:",BUP("abc").parse())
```
