```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv("iris.csv")[["sepal.length", "sepal.width"]]

print(f"Correlation Matrix:\n{df.corr()}")
print(f"\nCovariance Matrix:\n{df.cov()}")


plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.regplot(x=df.iloc[:,0], y=df.iloc[:,1])

# Heatmapy7
plt.subplot(1, 2, 2)
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", vmin=-1, vmax=1)

plt.tight_layout()
plt.show()
```
