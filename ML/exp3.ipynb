```python
# 3) Develop a program to Load a dataset with at least two numerical columns (e.g.,
# Iris, Titanic). Plot a scatter plot of two variables and calculate their Pearson correlation coefficient. 
# Generate a heatmap to visualize the correlation matrix of the numerical variables in the dataset. 
# Interpret the results and discuss any insights gained from the correlation analysis.

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv("iris.csv")[["sepal.length", "sepal.width"]]

print(f"Correlation Matrix:\n{df.corr()}")
print(f"\nCovariance Matrix:\n{df.cov()}")


plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.regplot(x=df.iloc[:,0], y=df.iloc[:,1])

# Heatmapy7
plt.subplot(1, 2, 2)
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", vmin=-1, vmax=1)

plt.tight_layout()
plt.show()
```
