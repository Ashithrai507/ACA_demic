```python

# linear regression:
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# 1. Load and Fit (rm: rooms, medv: price)
df = pd.read_csv("BostonHousing.csv")
X, y = df[['rm']], df['medv']
model = LinearRegression().fit(X, y)

# 2. Visualise
plt.scatter(X, y, alpha=0.5, label='Actual')
plt.plot(X, model.predict(X), color='red', label='Trend')
plt.title("Linear Regression: RM vs MEDV")
plt.legend()
plt.show()
print(f"R^2 Score: {model.score(X, y):.4f}")

```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1. Clean Data
df = pd.read_csv("auto-mpg.csv")
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
df = df.dropna(subset=['horsepower'])
X, y = df['horsepower'], df['mpg']

# 2. Polynomial Fit (Degree 2)
weights = np.polyfit(X, y, 2)
model = np.poly1d(weights)

# 3. Plot
xp = np.linspace(X.min(), X.max(), 100)
plt.scatter(X, y, alpha=0.5, color='green')
plt.plot(xp, model(xp), color='orange', lw=3)
plt.title("Polynomial Regression: Horsepower vs MPG")
plt.show()
print(f"Prediction for 100HP: {model(100):.2f}")
```
