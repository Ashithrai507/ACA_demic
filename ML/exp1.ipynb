```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.DataFrame({
    'Age':[23,23,27,27,39,41,47,49,50,52,54,54,56,57,58,58,60,61],
    'Fat':[7.8,9.5,17.8,25.9,26.5,27.2,27.4,28.8,30.2,31.2,31.4,32.9,33.4,34.1,34.6,35.7,41.2,42.5]
})
print("summary",df['Fat'].describe())
print("mode",df['Fat'].mode()[0])
df['Age group'] = pd.cut(df['Age'],bins=[0,30,50,100],labels=['young','medium','older'])

plt.figure(figsize=(10,4))
plt.subplot(1, 2, 1)
sns.histplot(df['Fat'], kde=True) 
plt.subplot(1, 2, 2)
sns.boxplot(x=df['Fat'])         
plt.show()

plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
df['Age group'].value_counts().plot(kind='bar', color='coral') # Bar Chart
plt.subplot(1, 2, 2)
df['Age group'].value_counts().plot(kind='pie', autopct='%1.1f%%') # Pie Chart
plt.show()
```
