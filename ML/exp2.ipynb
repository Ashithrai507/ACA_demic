```python

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import StandardScaler, LabelEncoder

df = pd.read_csv("iris.csv")
X = StandardScaler().fit_transform(df.iloc[:, :-1])
y = LabelEncoder().fit_transform(df.iloc[:, -1])
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)


for weight_type in ['uniform', 'distance']:
    print(f"\n--- KNN with {weight_type} weights ---")
    for k in [1, 3, 5]:
        knn = KNeighborsClassifier(n_neighbors=k, weights=weight_type)
        knn.fit(X_train, y_train)
        y_pred = knn.predict(X_test)
        
        print(f"K={k} Accuracy: {knn.score(X_test, y_test):.4f}")
        print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix

# 1. Setup & Preprocess
df = pd.read_csv("IRIS.csv")
X = StandardScaler().fit_transform(df.iloc[:, :-1])
y = LabelEncoder().fit_transform(df.iloc[:, -1])
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 2. Dual Loop Evaluation
for w in ['uniform', 'distance']:
    print(f"\n--- {w.capitalize()} k-NN ---")
    for k in [1, 3, 5]:
        knn = KNeighborsClassifier(n_neighbors=k, weights=w).fit(X_train, y_train)
        y_pred = knn.predict(X_test)
        
        acc, f1 = accuracy_score(y_test, y_pred), f1_score(y_test, y_pred, average='weighted')
        print(f"k={k} | Acc: {acc:.4f} | F1: {f1:.4f}")
        print(confusion_matrix(y_test, y_pred))
```
