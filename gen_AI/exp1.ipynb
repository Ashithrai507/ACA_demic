```python
# Install once (ignore if already installed)
# !pip install gensim

import gensim.downloader as api

# Load model (choose one)
model = api.load("glove-wiki-gigaword-50")

# Example 1: Direct similarity
print(model.most_similar(positive=['king', 'woman'], negative=['man'], topn=5))

# Example 2: Simple function for reuse
def word_relation(a, b, c):
    result = model.most_similar(positive=[a, c], negative=[b], topn=1)
    print(f"{a} - {b} + {c} → {result[0][0]}")

# Test
word_relation("king", "man", "woman")
```
